# Task 1: Image Cartoonifying Using Image Processing Filters

## Objective
Transform real-world images into cartoon-like images by applying various image processing techniques. The goal is to make flat regions more uniform and edges more distinct, creating a comic book effect.

## Cartoonifying Pipeline

### Step 1: Black & White Sketch Generation
- Convert image to grayscale (from BGR format)
- Apply **Median filter** for noise reduction while preserving edges
- Detect edges using **Laplacian filter**
- Apply **binary thresholding** to create clean sketch lines

### Step 2: Color Painting Generation
- Apply **bilateral filter** (edge-preserving smoothing) to:
  - Smooth flat regions
  - Maintain edge sharpness
- Optimize performance by:
  - Working at lower resolution
  - Applying multiple small bilateral filters instead of one large filter

### Step 3: Cartoon Creation
- Overlay the sketch (edge mask) onto the color painting
- Final output combines smooth colored regions with bold sketch lines

## Key Parameters
- **Median filter size**: Controls noise reduction
- **Laplacian filter**: Edge detection
- **Binary threshold**: Converts edges to black/white sketch lines
- **Bilateral filter parameters**:
  - Color strength
  - Positional strength
  - Filter size
  - Number of repetitions

## Expected Output
- **Sketch**: Black and white line drawing
- **Painting**: Smooth, poster-like color image
- **Cartoon**: Sketch overlaid on painting (final result)
